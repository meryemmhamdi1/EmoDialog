# Human Validation 

```python
emotion_annotated_transcripts = pd.read_csv('episodeslexicon.csv',escapechar='\\')
emotion_annotated_transcripts_bbt = emotion_annotated_transcripts[emotion_annotated_transcripts['SeriesName']=='BBT']
emotion_annotated_transcripts = emotion_annotated_transcripts[[u'Actor', u'MovieId', u'SeriesName',
       u'Subtitle', u'Emotionalities Lexicon Based',
       u'Emotion Categories Lexicon Based']]
emotion_annotated_transcripts.head()
```

```python
frequency_actors = Counter(emotion_annotated_transcripts_bbt['Actor'])
df = pd.DataFrame.from_dict(frequency_actors, orient='index')
df.sort_values([0], ascending=[0]).head(10)
```

```python
emotion_annotated_transcripts_emotional = emotion_annotated_transcripts[emotion_annotated_transcripts['Emotion Categories Lexicon Based']!='Neutral']
emotion_annotated_transcripts_emotional = emotion_annotated_transcripts_emotional[emotion_annotated_transcripts_emotional['Subtitle'].str.len()>200]
frequency_emotions = Counter(emotion_annotated_transcripts_emotional['Emotion Categories Lexicon Based'])
df = pd.DataFrame.from_dict(frequency_emotions, orient='index')
#df.sort_values([0], ascending=[0])
df.iloc[0]
frequency_emotions['Amusement-Laughter']
percentage_emotions = dict([(i, frequency_emotions[i] / float(len(emotion_annotated_transcripts_emotional['Emotion Categories Lexicon Based'])) * 100.0) for i in frequency_emotions])
percentage_emotions
```

```python
len(emotion_annotated_transcripts_emotional)
len(emotion_annotated_transcripts_emotional.iloc[0]['Subtitle'])
```

```python
length = []
for i in range(0,len(emotion_annotated_transcripts_emotional)):
    length.append(len(emotion_annotated_transcripts_emotional.iloc[i]['Subtitle']))
frequency_lengths = Counter(length)
df = pd.DataFrame.from_dict(frequency_lengths, orient='index')
df['Lengths']= df.index
df[df['Lengths']>200][0].sum()
```

```python
import random
sample_size = 250
rand_smpl = []
for emotion in percentage_emotions:
    emotion_annotated_transcripts_emotion = emotion_annotated_transcripts_emotional[emotion_annotated_transcripts_emotional['Emotion Categories Lexicon Based']==emotion]
    for i in sorted(random.sample(xrange(len(emotion_annotated_transcripts_emotion)), int(sample_size*percentage_emotions[emotion]/100))):
        rand_smpl.append(emotion_annotated_transcripts_emotion.iloc[i])
rand_smpl_df = pd.DataFrame(rand_smpl)
rand_smpl_df[['Subtitle','Emotion Categories Lexicon Based']]
```
