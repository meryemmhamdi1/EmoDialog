# Representative Emotional Patterns in Human Conversation

### Side Note: some types of responses to look at 
1. Reflective:  Responses that show an understanding of the thoughts and feelings that have been expressed by the student.
2. Interpretive:  Responses that indicate that the supervisor understands on a deeper level than the words spoken the meaning of the event or feeling that the student is explaining (reporting).  Seeks to explain the dynamics of a situation.
3. Supportive:  Responses that are intended to reassure and/or reduce anxiety. Responses that also can seek to inspire or to undergird a person.
4. Evaluative:  Responses that indicate that the supervisor has made a judgments regarding the relative quality, appropriateness, effectiveness or “rightness” of the students actions, words or behavior (use of responses that suggest what the student might or ought to have done). Can carry a value judgment.
5. Probing:  Responses that indicate that the supervisor’s intent is to query, to seek further information or provoke further discussion along a certain line.  Can be divided into probing for feelings and probing for information
6. Advising:  Responses that indicate the supervisors’ intention is to recommend certain approaches, actions, or methods of reflection and response. An offer of constructive suggestions about coping with a situation.
7. Teaching:  Responses that seek to impart information.

# Investigating Immitation Relationships between context/response detected emotions

```python
# Calculate probabilities with respect to each emotion in the context and its immitation in the response
from collections import Counter
immitation_matrix = []
for i in range(1,len(set(emotion_transcripts_source['emotion']))+1):
    probabilities = []
    emotion_source = emotion_transcripts_source[emotion_transcripts_source['emotion']==i]
    emotion_target = emotion_transcripts_target.loc[emotion_source.index.tolist()]
    frequency = Counter(emotion_target['emotion'])
    for j in range(1,21):
        if j not in frequency.keys():
            frequency[j] = 0
    sum_ = 0
    for key in frequency.keys():
        sum_ += frequency[key]
    for key in frequency.keys():
        probabilities.append((frequency[key]/ sum_)*100)
    immitation_matrix.append(probabilities[i-1]*(100/21))
```

### Investigating Polarity Relationships between context/response detected emotions:

```python
emo_polarity= {
    '1' :'Involvement-Interest',
    '2': 'Amusement-Laughter',
    '3': 'Pride-Elation',
    '4': 'Happiness-Joy',
    '5': 'Enjoyment-Pleasure',
    '6': 'Tenderness-Feeling Love',
    '7': 'Wonderment-Feeling Awe',
    '8': 'Feeling Disburdened- Relief', 
    '9': 'Astonishment- Surprise',
    '10': 'Longing- Nostalgia',
    '11': 'Pity-Compassion',
    '12': 'Sadness-Despair',
    '13': 'Worry-Fear',
    '14': 'Embarrassement-Shame',
    '15': 'Guilt-Remorse',
    '16': 'Disappointment- Regret',
    '17': 'Envy-Jealousy',
    '18': 'Disgust-Repulsion',
    '19': 'Contempt-Scorn',
    '20': 'Irritation-Anger',
    '21': 'Neutral'
}
for i in range(0,len(emotion_annotated_transcripts)):
    positive_emotions = [1,2,3,4,5,6,7,8,10]
    negative_emotions = list(range(11,21))
    neutral = 21
    if emotion_annotated_transcripts.loc[i]['EmotionId'] in positive_emotions:
        emotion_annotated_transcripts['Polarity'] = 'positive'
    elif emotion_annotated_transcripts.loc[i]['EmotionId'] in negative_emotions:
        emotion_annotated_transcripts['Polarity'] = 'negative'
    elif emotion_annotated_transcripts.loc[i]['EmotionId'] == neutral:
        emotion_annotated_transcripts['Polarity'] = 'neutral'
```

```python
from collections import Counter
polarity_matrix = []
for i in range(1,len(set(emotion_transcripts_source['EmotionId']))+1):
    probabilities = []
    emotion_source = emotion_transcripts_source[emotion_transcripts_source['EmotionId']==i]
    emotion_target = emotion_transcripts_target.loc[emotion_source.index.tolist()]
    frequency = Counter(emotion_target['EmotionId'])
    for j in range(1,21):
        if j not in frequency.keys():
            frequency[j] = 0
    sum_ = 0
    for key in frequency.keys():
        sum_ += frequency[key]
    # Calculate ratio of positive responses for that emotion
    sum_positive = 0
    for key in positive_emotions:
        sum_positive += frequency[key] 
    # Calculate ratio of negative responses for that emotion
    sum_negative = 0
    for key in negative_emotions:
        sum_negative += frequency[key] 
    # Calculate ratio of neutral responses for that emotion
    sum_neutral = frequency[neutral]
    probabilities.append((sum_positive/ sum_)*100)
    probabilities.append((sum_negative/ sum_)*100)
    probabilities.append((sum_neutral/ sum_)*100)
    polarity_matrix.append((probabilities)
```

```python
# Averaging over all the positive emotions
polarity_averages = []
pos_percent = 0
neg_percent = 0
neut_percent = 0
for i in positive_emotions:
    pos_percent += polarity_matrix[i][0]
    neg_percent += polarity_matrix[i][1]
    neut_percent += polarity_matrix[i][2]

pos_percent = pos_percent/10
neg_percent = neg_percent/10
neut_percent = neut_percent/10

polarity_averages.append([pos_percent,neg_percent,neut_percent])

# Averaging over all the negative emotions
pos_percent = 0
neg_percent = 0
neut_percent = 0
for i in negative_emotions:
    pos_percent += polarity_matrix[i][0]
    neg_percent += polarity_matrix[i][1]
    neut_percent += polarity_matrix[i][2]

pos_percent = pos_percent/10
neg_percent = neg_percent/10
neut_percent = neut_percent/10

polarity_averages.append([pos_percent,neg_percent,neut_percent])

polarity_averages
#polarity_matrix[0].index(max(polarity_matrix[0]))
```

```python
import glob
import re
import csv

import pandas as pd
header = ['Actor', 'Subtitle']
episodes = pd.DataFrame(columns=header)
allFiles = sorted(glob.glob('/home/meryem/Dropbox/meryem/Deliverables/1. Raw Dataset/BBT/BBT_Transcripts_Raw/*'))
for file_ in allFiles:
    df = pd.read_csv(file_
                     , sep='\s{5,}', engine='python', names=['col'])
    df = df[~ df['col'].str.startswith('(')]
    df['col'].str.replace(r"\(.*\)","")
    df = df[~ df['col'].str.startswith('Later')]
    df = df[~ df['col'].str.startswith('Credits sequence.')]
    df = df[df['col'].str.contains(':')]
    df['Actor'] = df['col'].str.split(r'[(]|:', 1).apply(lambda x: x[0])
    df['Subtitle'] = df['col'].str.split(r'[(]|:', 1).apply(lambda x: x[1])
    del (df['col'])

    #df = df[df.Actor != 'Scene']
    episodes = episodes.append(df)

episodes = episodes.reset_index(drop=True)
```

## Drawing Different Transition Matrices for Investigating Dialog Turns within Scenes

```python
from __future__ import division
emotion_annotated_transcripts_bbt_scene = emotion_annotated_transcripts_bbt[emotion_annotated_transcripts_bbt['Scene']==0]

emotion_transcripts_source_bbt_scene = emotion_annotated_transcripts_bbt_scene.iloc[::2].reset_index(drop=True)
emotion_transcripts_target_bbt_scene = emotion_annotated_transcripts_bbt_scene.iloc[1::2].reset_index(drop=True)

from collections import Counter
transition_matrix_context_scene = []
emotion_set_source = list(set(emotion_transcripts_source_bbt_scene['EmotionId']))
emotion_set_target = list(set(emotion_transcripts_target_bbt_scene['EmotionId']))
for i in emotion_set_source:
    probabilities = []
    emotion_source = emotion_transcripts_source_bbt_scene[emotion_transcripts_source_bbt_scene['EmotionId']==i]
    emotion_target = emotion_transcripts_target_bbt_scene.loc[emotion_source.index.tolist()]
    frequency = Counter(emotion_target['EmotionId'])
    for j in emotion_set_target:
        if j not in frequency.keys():
            frequency[j] = 0
    sum_ = 0
    for key in emotion_set_target:
        sum_ += frequency[key]
    for key in emotion_set_target:
        probabilities.append(float((frequency[key]/ sum_)*100))
    transition_matrix_context_scene.append(probabilities)
```
